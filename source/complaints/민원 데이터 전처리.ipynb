## 1. 전자민원처리부(2012~2023).xlsx

```python
import pandas as pd
import numpy as np
```

```python
complain1 = pd.read_excel("./data/1. 전자민원처리부(2012~2023).xlsx",header=1, sheet_name=None)
complain1
```

```python
complain_elec_2019 = complain1["2019년570건"]
complain_elec_2020 = complain1["2020년674건"]
complain_elec_2021 = complain1["2021년718건"]
complain_elec_2022 = complain1["2022년768건"]
complain_elec_2023 = complain1["2023년342건"]
```

```python
complain_elec_2019
```

```python
complain_elec_2019["민원요지"]
```

```python
from konlpy.tag import Okt

documents = [complain_elec_2019["민원요지"], complain_elec_2020["민원요지"], complain_elec_2021["민원요지"], complain_elec_2022["민원요지"], complain_elec_2023["민원요지"]]

# Okt 형태소 분석기 사용
okt = Okt()

# 텍스트 전처리 함수
def preprocess_text(text):
    # 형태소 분석
    tokens = okt.morphs(text, norm=True, stem=True)
    # 불용어 제거
    stopwords = ['에', '에서', '가', '을', '를', '이', '는', '은', '과', '와', '하다', '하는', '의', '년', '연', '도', '고','및','인','당','암','리','중','번지','대한','최근','', '(',')','*','*.','**.','***.','****.', '***-*', '**-*','*-*','***', '**','****','*.*','****-**','***-**','**-**','*-**','-','(*','*)','~','ㅣ','/','\\']
    tokens = [token for token in tokens if token not in stopwords]
    return tokens

okt_documents = pd.Series()

# 모든 문서에 대해 전처리 적용
for idx, document in enumerate(documents):
    
    processed_document = [preprocess_text(doc) for doc in document]
    okt_documents.loc[idx]=processed_document
okt_documents
```

```python
# 2019~2023 특수문자, 불용어(조사) 제외 민원 요지 토큰 데이터
okt_documents.to_csv("./tmp/okt_doc_elec.csv", index=False)
```

```python
documents = [complain_elec_2019["민원요지"], complain_elec_2020["민원요지"], complain_elec_2021["민원요지"], complain_elec_2022["민원요지"], complain_elec_2023["민원요지"]]

# 텍스트 전처리 함수
def preprocess_text(text):
    # split
    tokens = text.split()
    check = True
    newtokens = []

    stopwords = ['(',')','*','.','-','~','ㅣ','/','\\','']
    for token in tokens:
        for i in token:
            if i in stopwords:
                check = False
                break
        if check:
            newtokens.append(token)
        else:
            check = True
    return ' '.join(newtokens)

split_documents = pd.Series()

# 모든 문서에 대해 전처리 적용
for idx, document in enumerate(documents):
    
    processed_document = [preprocess_text(doc) for doc in document]
    split_documents.loc[idx]=processed_document
split_documents
```

```python
# 2019~2023 특수문자 제외 민원 요지 데이터
split_documents.to_csv("./tmp/split_doc_elec.csv", index=False)
```
